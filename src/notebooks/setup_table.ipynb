# Zerobus Demo - Table Setup
Creates the target schema, table, and grants for the Zerobus ingest demo.

```python
spark.sql("CREATE SCHEMA IF NOT EXISTS startups_catalog.dw_zerobus")
```

```python
spark.sql("""
  CREATE TABLE IF NOT EXISTS startups_catalog.dw_zerobus.zerobus_events (
    device_id STRING,
    ts TIMESTAMP,
    metric_value DOUBLE,
    status STRING
  )
""")
```

```python
service_principal = dbutils.widgets.get("service_principal_client_id")
if service_principal:
    spark.sql(f"GRANT USE CATALOG ON CATALOG startups_catalog TO `{service_principal}`")
    spark.sql(f"GRANT USE SCHEMA ON SCHEMA startups_catalog.dw_zerobus TO `{service_principal}`")
    spark.sql(f"GRANT MODIFY, SELECT ON TABLE startups_catalog.dw_zerobus.zerobus_events TO `{service_principal}`")
    print(f"Granted access to {service_principal}")
else:
    print("No service_principal_client_id provided, skipping GRANT.")
```
